```python
import numpy as np
import os
import json
import matplotlib.pyplot as plt
import seaborn as sb
from pathlib import Path
```

```python
Path("their_setup/old/output_plot/task3").mkdir(parents=True, exist_ok=True)
dir1 = os.path.abspath("their_setup/old")
task3_dir = os.path.join(dir1, "task3")
task3_sim_files = sorted([os.path.join(task3_dir, k) for k in os.listdir(os.path.join(task3_dir)) if "sim_matrix" in k and ".txt" in k])
for file in task3_sim_files:
    name = file.rsplit('/', 1)[-1]
    file_name = name.rsplit('.', 1)[0]
    print(file_name)
    sim_matrix = np.array(json.loads(json.load(open(file, "r"))))
    fig, ax = plt.subplots(figsize=(20, 20))
    sb.heatmap(sim_matrix)
    print(sim_matrix)
    fig1 = plt.gcf()
    plt.show()
    plt.draw()
    fig1.savefig('their_setup/old/output_plot/task3/' + file_name + '.png', dpi=100)
```
