```python
import pandas as pd
import os
```

```python
for f in [x for x in os.listdir(".") if ".csv" in x]:
    df = pd.read_csv(f)
    df = df.values.tolist()
    main = ""
    print("")
    for item in df:
        p = " & ".join([str(x) for x in item]) + " \\\\ \hline \n"
        main += p
    print(main)
```
