```python
d = pd.read_csv("Data_old/W/3.csv", header=None)
seq = d.iloc[0,:].values
plt.figure(figsize=(8,5))
for i in range(0,len(seq),3):
    plt.axvline(i, linestyle="--", color="green")
    val = np.round(np.mean(seq[i:i+3]),decimals=2)
    if i%2 == 0:
        plt.text((i+i+3)/2-1.5, val, str(val), color="blue", fontsize=15)
plt.plot(seq, color="red")
plt.xlabel("time", fontsize=15)
plt.ylabel("sensor values", fontsize=15)
plt.savefig("task0_1.png")

d = pd.read_csv("Data_old/W/15.csv", header=None)
seq = d.iloc[0,:].values
plt.figure(figsize=(8,5))
for i in range(0,len(seq),3):
    plt.axvline(i, linestyle="--", color="green")
    val = np.round(np.mean(seq[i:i+3]),decimals=2)
    if i%2 == 0:
        plt.text((i+i+3)/2-1.5, val, str(val), color="blue", fontsize=15)
plt.plot(seq, color="red")
plt.xlabel("time", fontsize=15)
plt.ylabel("sensor values", fontsize=15)
plt.savefig("task0_2.png")

d = pd.read_csv("Data_old/W/24.csv", header=None)
seq = d.iloc[0,:].values
plt.figure(figsize=(8,5))
for i in range(0,len(seq),3):
    plt.axvline(i, linestyle="--", color="green")
    val = np.round(np.mean(seq[i:i+3]),decimals=2)
    if i%2 == 0:
        plt.text((i+i+3)/2-1.5, val, str(val), color="blue", fontsize=15)
plt.plot(seq, color="red")
plt.xlabel("time", fontsize=15)
plt.ylabel("sensor values", fontsize=15)
plt.savefig("task0_3.png")

d = pd.read_csv("Data_new_test/W/24.csv", header=None)
seq = d.iloc[0,:].values
plt.figure(figsize=(8,5))
for i in range(0,len(seq),3):
    plt.axvline(i, linestyle="--", color="green")
    val = np.round(np.mean(seq[i:i+3]),decimals=2)
    if i%2 == 0:
        plt.text((i+i+3)/2-1.5, val, str(val), color="blue", fontsize=15)
plt.plot(seq, color="red")
plt.xlabel("time", fontsize=15)
plt.ylabel("sensor values", fontsize=15)
plt.savefig("task0_a.png")

d = pd.read_csv("Data_new_test/W/278.csv", header=None)
seq = d.iloc[0,:].values
plt.figure(figsize=(8,5))
for i in range(0,len(seq),3):
    plt.axvline(i, linestyle="--", color="green")
    val = np.round(np.mean(seq[i:i+3]),decimals=2)
    if i%2 == 0:
        plt.text((i+i+3)/2-1.5, val, str(val), color="blue", fontsize=15)
plt.plot(seq, color="red")
plt.xlabel("time", fontsize=15)
plt.ylabel("sensor values", fontsize=15)
plt.savefig("task0_b.png")

d = pd.read_csv("Data_new_test/W/575.csv", header=None)
seq = d.iloc[0,:].values
plt.figure(figsize=(8,5))
for i in range(0,len(seq),3):
    plt.axvline(i, linestyle="--", color="green")
    val = np.round(np.mean(seq[i:i+3]),decimals=2)
    if i%2 == 0:
        plt.text((i+i+3)/2-1.5, val, str(val), color="blue", fontsize=15)
plt.plot(seq, color="red")
plt.xlabel("time", fontsize=15)
plt.ylabel("sensor values", fontsize=15)
plt.savefig("task0_c.png")
```

```python
from pathlib import Path
import pandas as pd
import json
import os
import numpy as np

dir = "old_theirs_nonNormalized"
Path(dir+"/output_plot/task2").mkdir(parents=True, exist_ok=True)
data = {}
cols = []
for f in os.listdir(dir+"/task2"):
    d = json.load(open(dir+"/task2/"+f))
    file = f.split("_")[-2]
    vm = f.split("_")[-1].split(".")[0]
    op = "_".join(f.split("_")[:2])
    if op not in cols:
        cols.append(op)
    if file not in data:
        data[file] = {}
    if vm not in data[file]:
        data[file][vm] = {}
    data[file][vm][op] = list(d.keys())

for f in data:
    for vm in data[f]:
        df = pd.DataFrame(np.array(list(data[f][vm].values())).T, columns=cols)
        df.to_csv(dir+"/output_plot/task2/{}_{}.csv".format(f,vm),sep=",",index=None)
```
